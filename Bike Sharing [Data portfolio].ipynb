```python
# Defining a dictionary (for easy namimg)
new_cols_dict = {
'timestamp':'time',
'cnt':'count',
't1':'temp_real_C',
't2':'temp_feels_like_C',
'hum':'humidity_percent',
'wind_speed':'wind_speed_kph',
'weather_code': 'weather',
'is_holiday':'is_holiday',
'is_weekend':'is_weekend',
'season':'season'
}

# Renaming columns
bikes.rename(new_cols_dict, axis = 1, inplace = True)
```

```python
# Changing the humidity values to a value from 100%
bikes.humidity_percent = bikes.humidity_percent/100
```

```python
# Checking values to see if the data is correct
bikes.humidity_percent[1:10,]
```

```python
# Making a seasons and weather dictionary
seasons_dict = {
'0.0': 'spring',
'1.0': 'summer',
'2.0': 'autumn',
'3.0': 'winter'
}

weather_dict = {
'1.0':'Clear',
'2.0': 'Scattered Clouds',
'3.0': 'Broken clouds',
'4.0': 'Cloudy',
'7.0': 'Rain',
'10.0': 'Rain with thunderstorm',
'26.0': 'Snowfall'
}
```

```python
# Manipulating vlaues of the column(s)
bikes.season = bikes.season.astype('str')

# Using map function
bikes.season = bikes.season.map(seasons_dict)

# Weather
bikes.weather = bikes.weather.astype('str')

# Mapping function to replace
bikes.weather = bikes.weather.map(weather_dict)
```

```python
# Checking to see if data has been changed successfully
bikes.head()
```

```python
# Mapping the is_holiday and _is.weekend fields
holiday_dict = {
'0.0':'No',
'1.0':'Yes'
}

weekend_dict = {
'0.0':'No',
'1.0':'Yes'
}


# Revising the data attributes for mapping
bikes.is_holiday = bikes.is_holiday.astype('str')
bikes.is_weekend = bikes.is_weekend.astype('str')

# Mapping
bikes.is_holiday = bikes.is_holiday.map(holiday_dict)
bikes.is_weekend = bikes.is_weekend.map(weekend_dict)
```

```python
# Finalising data manipulation and creating excel sheet for Tableau visualisation

try:
    bikes.to_excel('london_bikes_final.xlsx', sheet_name='Data')
    print("File written successfully.")
except FileExistsError:
    # Handle the case where the file already exists
    print("File already exists. It will be replaced.")
```
